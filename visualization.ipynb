```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wer_df = pd.read_csv("wer_pretrained_vs_non_pretrained.csv")
cer_df = pd.read_csv("cer_pretrained_vs_non_pretrained.csv")
loss_df = pd.read_csv("loss_pretrained_vs_non_pretrained.csv")


loss_df.head()
```

```python
wer_pretrain_col = "[only ctc] wav2vec+pretrained_dns64 [lr=1e-4] (unfreeze), SNR=10, train[0:20000] second run - val/wer (gt-denoisy)"
wer_non_pretrain_col = "RU [only ctc] wav2vec+pretrained_dns64 [lr=1e-4] (non pretrained), SNR in [10, 10], train[0:20000] - val/wer (gt-denoisy)"

cer_pretrain_col = "[only ctc] wav2vec+pretrained_dns64 [lr=1e-4] (unfreeze), SNR=10, train[0:20000] second run - val/cer (gt-denoisy)"
cer_non_pretrain_col = "RU [only ctc] wav2vec+pretrained_dns64 [lr=1e-4] (non pretrained), SNR in [10, 10], train[0:20000] - val/cer (gt-denoisy)"

loss_pretrain_col = "[only ctc] wav2vec+pretrained_dns64 [lr=1e-4] (unfreeze), SNR=10, train[0:20000] second run - val/asr_loss"
loss_non_pretrain_col = "RU [only ctc] wav2vec+pretrained_dns64 [lr=1e-4] (non pretrained), SNR in [10, 10], train[0:20000] - val/asr_loss"
```

```python
# Создание субплотов
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# График для первого датафрейма
axes[0].plot(wer_df['epoch'], wer_df[wer_pretrain_col], label='pretrained', marker='o')
axes[0].plot(wer_df['epoch'], wer_df[wer_non_pretrain_col], label='not-pretrainded', marker='x')
axes[0].legend()
axes[0].set_title('Validation WER')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('WER')
axes[0].set_xlim([0, 18])

# График для первого датафрейма
axes[1].plot(cer_df['epoch'], cer_df[cer_pretrain_col], label='pretrained', marker='o')
axes[1].plot(cer_df['epoch'], cer_df[cer_non_pretrain_col], label='not-pretrainded', marker='x')
axes[1].legend()
axes[1].set_title('Validation CER')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('CER')
axes[1].set_xlim([0, 18])

# График для первого датафрейма
axes[2].plot(loss_df['epoch'], loss_df[loss_pretrain_col], label='pretrained', marker='o')
axes[2].plot(loss_df['epoch'], loss_df[loss_non_pretrain_col], label='not-pretrainded', marker='x')
axes[2].legend()
axes[2].set_title('Validation CTC-loss')
axes[2].set_xlabel('Epoch')
axes[2].set_ylabel('CTC-loss')
axes[2].set_xlim([0, 18])

# Автоматическая подгонка параметров макета
plt.tight_layout()

# Отображение графиков
plt.show()
```
